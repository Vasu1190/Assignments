q.3)BMI Enter the BMI Index :34 ---Very Overweight


```python
BMI=int(input('Enter the BMI Index:'))
if(BMI<18):
    print('Underweight')
elif(BMI<=25):
    print('Normal Weight')
elif(BMI<=30):
    print('Overweight')
elif(BMI<=40):
    print('Very Overweight')
else:
    print('Severely Overweight')
```
