# Merging 2 datasets one named Materials and their Mechanical Properties & the other dataset named Use. 

```python
dataset1= pd.read_csv('Materials and their Mechanical Properties.csv')
dataset2=pd.read_csv('Use.csv')
```

```python
# Assuming 'dataset1' is your first dataset
dataset1['ID'] = range(1, len(dataset1) + 1)


# Assuming 'dataset2' is your second dataset
dataset2['ID'] = range(1, len(dataset2) + 1)

# Merge dataset 1 and dataset 2 using the 'ID' column
dataset = dataset1.merge(dataset2[['ID', 'Use']], on='ID', how='inner')
dataset
```

```python
dataset['Materials'] = dataset['Std'] + ' ' + dataset['Material'] + ' ' + dataset['Heat treatment']

# Drop the original columns
dataset = dataset.drop(['Std', 'Material', 'Heat treatment'], axis=1)
dataset.head()
```

# Checking for text values in the numerical column before converting it to numeric datatype

```python
numerical_columns = ['Su', 'Sy', 'E', 'G', 'mu', 'Ro']

# Initialize an empty list to store column names with text values
columns_with_text = []

# Iterate through the numerical columns and check for text values
for column in numerical_columns:
    if dataset[column].apply(lambda x: isinstance(x, str)).any():
        columns_with_text.append(column)

# Display the columns with text values
print("Columns with text values:", columns_with_text)
```

```python
dataset.info()
```

```python
# Convert the entire column to numeric, setting errors='coerce' for non-numeric values
dataset['Sy'] = pd.to_numeric(dataset['Sy'], errors='coerce')

# Create a mask to identify rows with non-numeric values
mask = np.isnan(dataset['Sy'])

# Remove rows with non-numeric (categorical) values
dataset = dataset[~mask]
```

#  Copying the cleaned data into a new DataFrame

```python
dataset=dataset.copy()
dataset
```

```python
# Define a function to clean and format the material names
def clean_material(Material):
    # Remove special characters and digits
    cleaned_material = re.sub(r'[^a-zA-Z0-9 - ]', '', Material)
    # Remove extra spaces and convert to title case
    cleaned_material = ' '.join(cleaned_material.split())
    return cleaned_material

# Apply the function to the 'Material' column
dataset['Material'] = dataset['Material'].apply(clean_material)

dataset
```

# Defining Quan & Qual for Outlier Removal

```python
def quanQual(dataset):
    quan=[]
    qual=[]
    for columnName in dataset.columns:
        #print(columnName)
        if(dataset[columnName].dtype=='O'):
            #print("qual")
            qual.append(columnName)
        else:
            #print("quan")
            quan.append(columnName)
    return quan,qual
```

```python
quan,qual=quanQual(dataset)
quan
```

```python
qual
```

```python
descriptive=pd.DataFrame(index=["Mean","Median","Mode","Q1:25%","Q2:50%",
                               "Q3:75%","99%","Q4:100%","IQR","1.5rule","Lesser","Greater","Min","Max"],columns=quan)
for columnName in quan:
    descriptive[columnName]["Mean"]=dataset[columnName].mean()
    descriptive[columnName]["Median"]=dataset[columnName].median()
    descriptive[columnName]["Mode"]=dataset[columnName].mode()[0]
    descriptive[columnName]["Q1:25%"]=dataset.describe()[columnName]["25%"]
    descriptive[columnName]["Q2:50%"]=dataset.describe()[columnName]["50%"]
    descriptive[columnName]["Q3:75%"]=dataset.describe()[columnName]["75%"]
    descriptive[columnName]["99%"]=np.percentile(dataset[columnName],99)
    descriptive[columnName]["Q4:100%"]=dataset.describe()[columnName]["max"]
    descriptive[columnName]["IQR"]=descriptive[columnName]["Q3:75%"]-descriptive[columnName]["Q1:25%"]
    descriptive[columnName]["1.5rule"]=1.5*descriptive[columnName]["IQR"]
    descriptive[columnName]["Lesser"]=descriptive[columnName]["Q1:25%"]-descriptive[columnName]["1.5rule"]
    descriptive[columnName]["Greater"]=descriptive[columnName]["Q3:75%"]+descriptive[columnName]["1.5rule"]
    descriptive[columnName]["Min"]=dataset[columnName].min()
    descriptive[columnName]["Max"]=dataset[columnName].max()
```

```python
lesser=[]
greater=[]

for columnName in quan:
    if(descriptive[columnName]["Lesser"]>descriptive[columnName]["Min"]):
        lesser.append(columnName)
    if(descriptive[columnName]["Greater"]<descriptive[columnName]["Q4:100%"]):
        greater.append(columnName)
```
