# Metric 1:Retrieval Accuracy of a Chatbot Based On Natural Language Inference (NLI).

```python
import torch
from sentence_transformers import SentenceTransformer, util

# Load the pre-trained Sentence Transformer model (e.g., Sentence-BERT)
model = SentenceTransformer("all-mpnet-base-v2")

def preprocess_text(text):
    """Preprocesses text by lowering it and removing punctuation."""
    text = text.lower()
    text = ''.join([c for c in text if c.isalnum() or c.isspace()])
    return text

def evaluate_chatbot_nli(chat_history, test_data):
    total_questions = len(test_data)
    correct_responses = 0

    for user_query, expected_response in test_data:
        user_query_processed = preprocess_text(user_query)
        expected_response_processed = preprocess_text(expected_response)

        user_query_embedding = model.encode(user_query_processed)
        expected_response_embedding = model.encode(expected_response_processed)

        bot_responses = [conversation[1] for conversation in chat_history if user_query.lower() in conversation[0].lower()]
        bot_responses_processed = [preprocess_text(response) for response in bot_responses]

        # Ensure model outputs PyTorch tensors (check model documentation if needed)
        if not isinstance(user_query_embedding, torch.Tensor):
            user_query_embedding = torch.tensor(user_query_embedding)
        if not isinstance(expected_response_embedding, torch.Tensor):
            expected_response_embedding = torch.tensor(expected_response_embedding)

        bot_response_embeddings = model.encode(bot_responses_processed)  # Encode bot responses
        if not isinstance(bot_response_embeddings, torch.Tensor):  # Convert if necessary
            bot_response_embeddings = torch.tensor(bot_response_embeddings)

        # Check if there are any bot responses before similarity calculation
        if bot_response_embeddings.shape[0] > 0:
            similarities = util.pytorch_cos_sim(expected_response_embedding.unsqueeze(0), bot_response_embeddings)
            most_similar_index = similarities.argmax()
            most_similar_score = similarities.max().item()

        # Consider a response with a cosine similarity above a threshold as a match
            if most_similar_score > 0.3:
                correct_responses += 1
            else:
                print("User Query:", user_query)
                print("Expected Response:", expected_response_processed)
                print("Most Similar Bot Response (", most_similar_score, "):")
                if bot_response_embeddings.shape[0] > 0 and len(bot_responses_processed) > 0:
                    print(bot_responses_processed[most_similar_index])
                else:
                    print("No matching bot responses found for user query:", user_query)
        else:
            print("No matching bot responses found for user query:", user_query)


    accuracy = correct_responses / total_questions
    return accuracy

# Evaluate the chatbot
accuracy = evaluate_chatbot_nli(chat_history, test_data)
print("Accuracy:", accuracy)
```

# Metric 2: Conversational Relevance

```python
def is_conversationally_relevant(user_query, bot_response, weight_keywords=0.7, weight_conversation_flow=0.3):
    """
    Assesses the conversational relevance of the bot's response with a weighted score.

    Args:
        user_query: The user's query.
        bot_response: The chatbot's response.
        weight_keywords: Weighting factor for keyword matching (default: 0.7).
        weight_conversation_flow: Weighting factor for conversation flow (default: 0.3).

    Returns:
        A float value between 0 and 1 representing the relevancy score.
    """
    # Lowercase both strings for case-insensitive comparison
    user_query_lower = user_query.lower()
    bot_response_lower = bot_response.lower()

    # 1. Keyword Matching:
    keywords_in_response = any(keyword in bot_response_lower for keyword in user_query_lower.split())
    keyword_score = weight_keywords if keywords_in_response else 0

    # 2. Conversation Flow:
    chat_history = []  # This should be passed as an argument if needed, not defined inside
    chat_history_size = 2  # Adjust this value as needed

    if len(chat_history) > chat_history_size:
        chat_history.pop(0)  # Remove oldest entry if history is full

    chat_history.append(user_query_lower)
    mentions_previous_topic = any(keyword in bot_response_lower for keyword in chat_history)
    conversation_flow_score = weight_conversation_flow if mentions_previous_topic else 0

    # Calculate and return the weighted relevancy score
    relevancy_score = keyword_score + conversation_flow_score
    return relevancy_score

def evaluate_conversation_relevance(chat_history):
    """
    Evaluates the overall conversational relevance of the chatbot's responses.

    Args:
        chat_history: A list of tuples containing user queries and bot responses.

    Prints:
        The overall number of relevant and not relevant responses based on a threshold.
    """
    total_relevant = 0
    total_not_relevant = 0
    threshold = 0.5  # Adjust threshold for relevancy classification

    for user_query, bot_response in chat_history:
        score = is_conversationally_relevant(user_query, bot_response)
        if score >= threshold:
            total_relevant += 1
            print("Relevant Response:", bot_response, " (Score:", score, ")")
        else:
            total_not_relevant += 1
            print("Not Relevant:", bot_response, " (Score:", score, ")")

    print("\nOverall:")
    print("Relevant:", total_relevant)
    print("Not Relevant:", total_not_relevant)



evaluate_conversation_relevance(chat_history)
```

```python
def rate_fluency_coherence(response, scale=5):
    """
    Assesses the fluency and coherence of a chatbot response using a basic rule-based approach.

    Args:
        response: The chatbot's response.
        scale: The scale for the rating (default: 5).

    Returns:
        A float value between 0 and the scale representing the fluency and coherence score.
    """

    # Sentence Length and Structure: Penalize for very short or very long sentences.
    num_words = len(response.split())
    sentence_length_penalty = 0
    if num_words < 5:
        sentence_length_penalty = 0.5  # Adjust penalty as needed
    elif num_words > 20:
        sentence_length_penalty = 1.0  # Adjust penalty as needed

    # Grammatical Completeness: Check for basic structure and penalize for errors.
    grammatical_errors = 0
    if not any(verb in response for verb in ["is", "are", "was", "were", "has", "have", "do", "does", "did"]):
        grammatical_errors = 1.0  # Adjust penalty as needed

    # Word Choice and Clarity: Penalize for complex or nonsensical vocabulary.
    vocabulary_errors = 0 

    # Combine penalties for a weighted score
    fluency_coherence_score = scale - (sentence_length_penalty + grammatical_errors + vocabulary_errors)
    return fluency_coherence_score

def evaluate_fluency_coherence(responses, scale=5):
    """
    Evaluates the fluency and coherence of chatbot responses.

    Args:
        responses: A list of chatbot responses.
        scale: The scale for the rating (default: 5).

    Prints:
        The fluency and coherence score for each response.
    """

    fluency_coherence_scores = []
    for response in responses:
        score = rate_fluency_coherence(response, scale)
        fluency_coherence_scores.append(score)
        print("Fluency & Coherence Score:", score, "for response:", response)

    average_score = sum(fluency_coherence_scores) / len(fluency_coherence_scores)
    print("\nAverage Fluency and Coherence Score:", average_score)


evaluate_fluency_coherence(responses)
```

```python
import spacy

nlp = spacy.load("en_core_web_sm")

def evaluate_chatbot_dependency_parsing(chat_history, test_data):
    total_questions = len(test_data)
    correct_responses = 0

    for user_query, expected_response in test_data:
        expected_response_processed = nlp(expected_response)
        expected_dependencies = set([(token.dep_, token.text) for token in expected_response_processed])

        for conversation in chat_history:
            if user_query.lower() in conversation[0].lower():
                bot_response = conversation[1]
                bot_response_processed = nlp(bot_response)
                bot_dependencies = set([(token.dep_, token.text) for token in bot_response_processed])

                # Consider a response with a high number of matching dependencies as a match (adjust threshold)
                if len(expected_dependencies.intersection(bot_dependencies)) / len(expected_dependencies) > 0.8:
                    correct_responses += 1
                    break  # Stop searching after finding a match

    accuracy = correct_responses / total_questions
    return accuracy

# Evaluate the chatbot
accuracy = evaluate_chatbot_dependency_parsing(chat_history, test_data)
print("Accuracy:", accuracy)
```
